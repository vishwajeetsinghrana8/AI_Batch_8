```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
X = np.arange(-5, 5, 0.25) 
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X,Y,Z, rstride=1, cstride=1)
plt.show()
```

# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0.0,6.0)
y1 = np.sin(2*np.pi*x1)

x2 = np.linspace(0.0,6.0)
y2 = np.cos(2*np.pi*x2)*np.exp(-x2)
```

```python
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.subplot(2,1,2)
plt.plot(x2,y2,'o-')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Bar

```python
x1 = [1,2,3,4,5]
y1 = [10,24,36,45,39]

label = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

plt.bar(x1,y1, tick_label=label, width=0.6, color=['g','r','b','y'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.show()
```

# Histogram

```python
ages = [2,6,7,10,40,70,30,35,50,45,70,
       77,80,88,76,18,81,34,22,25]

range =(0,100)
bins = 10

plt.hist(ages, bins, range, color='g', histtype='bar', rwidth=0.6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
plt.show()
```
